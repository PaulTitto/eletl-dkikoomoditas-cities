### Data Profiling Method

```python
n_rows = data.shape[0]
n_cols = data.shape[1]

print(n_rows,n_cols)
```

```python
def percent_missing_values(col_name: str) -> float:
    n_data = len(data)
    n_missing_data = data[col_name].isnull().sum()
    percent_value = n_missing_data * 100 / n_data
    return percent_value
```

```python
print(data.columns)
```

```python
missing_values_dict = {}

for column in data.columns:
    missing_values_dict[column] = percent_missing_values(column)
```
